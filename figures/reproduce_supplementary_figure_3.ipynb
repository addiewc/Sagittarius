# Python notebook to reproduce _Supplementary Figure 3_

```python
import pandas as pd
import numpy as np
import os
import matplotlib.pyplot as plt

import plot_settings
import plot_utils
```

```python
fig_dir = 'supplementary/figure_3/'
if not os.path.exists(fig_dir):
    os.makedirs(fig_dir)
```

```python
df = pd.read_csv('results/LINCS/processed_dataset_statistics.csv')
```

# Code to reproduce supplementary figure 3a,b

Supplementary Fig. 3a

```python
%matplotlib inline

cl_map = {cl: i for i, cl in enumerate(np.unique(df['cell_id'].tolist()))}
dr_map = {dr: i for i, dr in enumerate(np.unique(df['drug_id'].tolist()))}
obs = np.zeros((len(cl_map), len(dr_map)))
for idx, row in df.iterrows():
    obs[cl_map[row['cell_id']], dr_map[row['drug_id']]] = 1
    
ax = plot_settings.get_wider_axis()
plot_utils.show_image(ax, obs, 'Drug', 'Cell line', aspect='auto', cmap='magma')
plot_utils.format_ax(ax)
plt.savefig(fig_dir + 'figure_3a.pdf')
```

Supplementary Fig. 3b

```python
ax = plot_settings.get_wider_axis()
plot_utils.histogram(ax, df['num_points'], 'Measured sequence length', 'Count', N_bins=25, yscale='log',
                     color=plot_settings.get_base_color(), edge_color=plot_settings.get_model_colors('cvae'))
plot_utils.format_ax(ax)
plt.savefig(fig_dir + 'figure_3b.pdf')
```
