# Notebook to reproduce _Supplementary Figure 19_

```python
import pandas as pd
import numpy as np
import json
import matplotlib.pyplot as plt
import os

import plot_utils
import plot_settings
```

```python
fig_dir = 'supplementary/figure_19/'

if not os.path.exists(fig_dir):
    os.makedirs(fig_dir)
```

# Produce supplementary figure 19a,b

Supplementary Fig. 19a

```python
%matplotlib inline

with open('results/TCGA/THCA_survival_time_and_grade.txt', 'r') as f:
    res = json.load(f)

%matplotlib inline

ax = plot_settings.get_wider_axis()
colors = [plot_settings.get_sag_vs_baseline_colors(m) for m in ['base', 'Sagittarius']]

plot_utils.kaplan_meier_curve(ax, res['survival time'], res['obs'],
                              ['AJCC Stage 1-3', 'AJCC Stage 4'], xlabel='Time (months)',
                              ylabel='Survival probability', showCI=False, colors=colors)
plt.title('THCA')
plot_utils.format_ax(ax)
plt.savefig(fig_dir + 'figure_19a.pdf', bbox_inches='tight')
```

Supplementary Fig. 19b

```python
%matplotlib inline

with open('results/TCGA/SARC_survival_time_and_grade.txt', 'r') as f:
    res = json.load(f)

%matplotlib inline

ax = plot_settings.get_wider_axis()
colors = [plot_settings.get_sag_vs_baseline_colors(m) for m in ['base', 'Sagittarius']]

plot_utils.kaplan_meier_curve(ax, res['survival time'], res['obs'],
                              ['FNCLCC grades 1-2', 'FNCLCC grade 3'], xlabel='Time (months)',
                              ylabel='Survival probability', showCI=False, colors=colors)
plt.title('SARC')
plot_utils.format_ax(ax)
plt.savefig(fig_dir + 'figure_19b.pdf', bbox_inches='tight')
```
